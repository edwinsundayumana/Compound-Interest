```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
P = 450000
T = [i for i in range(1,100)]
```

```python
def function(F,R,T):
    for i in T:
        f = P * (1+ R/100)**i
        F.append(f)
        result = pd.DataFrame(F, columns = "Compound_Interest".split())
        result.index.rename("Year")
    return result
```

```python
A, B, C, D, E = [P], [P], [P], [P], [P]
```

```python
R = [0.5, 1.5, 2.5, 3.5, 4.5]
```

```python
for r in R:
    if r == 0.5:
        A = function(A, r, T)
    elif r == 1.5:
        B = function(B, r, T)
    elif r == 2.5:
        C = function(C, r, T)
    elif r == 3.5:
        D = function(D, r, T)
    else:
        E = function(E, r, T)
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,10))

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 


ax.plot(A, label="Rate=0.5")
ax.plot(B, label="Rate=1.5")
ax.plot(C, label="Rate=2.5")
ax.plot(D, label="Rate=3.5")
ax.plot(E, label="Rate=4.5")
plt.xlabel("No of Years")
plt.ylabel("Compound Interest")
ax.yaxis.set_major_formatter(formatter) 
plt.title("A Graph Showing the Compound Interest")
plt.legend(loc="best")
```

```python
df = pd.concat([A,B,C,D,E], axis=1)
```
